# Which are the sentences where "love" was used?

```python
pattern = re.compile("[A-Z]{1}[^.]*[^a-zA-Z]+love[^a-zA-Z]+[^.]*.")
"""
    "[A-Z]{1}": Matches exactly one uppercase letter (A-Z).
    [^.]*: Matches zero or more characters that are not a period (.).
    [^a-zA-Z]+: Matches one or more characters that are not letters (neither uppercase nor lowercase).
    "love": Matches the exact word "love".
    [^a-zA-Z]+: Ensures that "love" is surrounded by non-letter characters.
    [^.]*: Matches zero or more characters that are not a period after "love".
    .: Matches a literal period (.) to denote the end of a sentence.
"""
finding = re.findall(pattern, book)
finding
```

# What are the most used words?

```python
pattern = re.compile("[a-zA-Z]+")
print(pattern)
findings = re.findall(pattern, book.lower())

len(findings)
findings
```

```python
d = {}
for word in findings:
    if word in d.keys():
        d[word] = d[word] + 1
    else:
        d[word] = 1
d
```

```python
d_list = []
for (key, value) in d.items():
    d_list.append((value, key))
sorted(d_list, reverse=True)
```

# Extract the paragraphs where "love" was used

```python
pattern = re.compile("[A-Z]{1}[^\r\n]*[^a-zA-Z]+love[^a-zA-Z]+[^\r\n]*")
finding = re.findall(pattern, book)
finding
```

```python

words = book.split()

# Create an empty dictionary to store word counts
word_counts = {}

# Iterate through each word
for word in words:
    # Remove any punctuation marks from the word
    word = word.strip(".,!?;:'\"")
    # Convert the word to lowercase to ensure case-insensitive counting
    word = word.lower()
    # Increment the count for the word in the dictionary
    word_counts[word] = word_counts.get(word, 0) + 1

# Print the count for each word
for word, count in word_counts.items():
    print(f"{word}: {count}")
```

# Fucntion that finds the occurrence of any word

```python
pattern = re.compile("[a-zA-Z]+")
findings = re.findall(pattern, book.lower())
d = {}
for word in findings:
    if word in d.keys():
        d[word] = d[word] + 1
    else:
        d[word] = 1
def find(w):
    try:
        return d[w]
    except:
        return f'The book does not contain the word "{w}"'
```
